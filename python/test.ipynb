```python
import numpy as np
a = [1,2,3,6,20]

size = len(a)

f = [[np.inf for x in range(256)] for y in range(size)]

def sqr(x):
	return x*x
def min(x,y):
	return x if x < y else y
def max(x,y):
	return x if x > y else y

for i in range(256):
	f[0][i] = sqr(i-a[0])

for i in range(1, size):
	for j in range(256):
		cost_j = sqr(a[i] - j)
		for k in range(max(0, j-3), j+1):
			f[i][j] = min(f[i][j], f[i-1][k] + cost_j)

ans = np.inf
for i in range(256):
    print (i, f[-1][i])
    ans = min(ans, f[-1][i])
print (ans)
```

```python
T = int(input())
while T > 0:
    T -= 1
    m = int(input())
    ans = 0
    if m > 0:
        ans = 1
        features = list(map(int, input().split()))
        old_dict = {}
        for i in range(1, len(features), 2):
            old_dict[(features[i], features[i+1])] = 1
        for i in range(1, m):
            features = list(map(int, input().split()))
            new_dict = {}
            for j in range(1, len(features), 2):
                pair = (features[j], features[j+1])
                new_dict[pair] = old_dict.get(pair, 0) + 1
                if new_dict[pair] > ans:
                    ans = new_dict[pair]
            old_dict.clear()
            old_dict = new_dict
    print (ans)
```
